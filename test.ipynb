```python
%%sql
SELECT *, ("High Price" - "Low Price") / 2 + "Low Price" AS "Midpoint Price" FROM us_pumpkin_prices LIMIT 5;
```

```python
%%sql
SELECT COUNT(*)
FROM us_pumpkin_prices;
```

```python
%%sql
SELECT COUNT("Item Size")
FROM us_pumpkin_prices;
```

```python
%%sql 
SELECT "Variety", COUNT(*) FROM us_pumpkin_prices GROUP BY "Variety";
```

```python
%%sql
SELECT "City Name", "Variety", COUNT(*) 
FROM us_pumpkin_prices 
GROUP BY "City Name", "Variety" 
HAVING MAX("High Price") > 300 
ORDER BY "City Name" ASC
```

```python
%%sql
SELECT "City Name", AVG("High Price" - "Low Price") AS "Difference"
FROM us_pumpkin_prices 
GROUP BY "City Name"
ORDER BY "Difference" DESC
```
